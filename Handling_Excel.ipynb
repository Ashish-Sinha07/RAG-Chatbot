## Handing the Excel File using RAG

```python
import pandas as pd
import seaborn as sns
import google.generativeai as genai
import os

# from langchain_core.documents import Document
```

```python
data = sns.load_dataset('tips')
```

```python
data.head(2)
```

```python
# LLM Model
gemini_key = os.getenv('GEMINI-API-KEY-3')
genai.configure(api_key=gemini_key)
model = genai.GenerativeModel('gemini-2.5-flash-lite')

```

```python
# Extract rows from excel and put them serially in text
context = ''
for idx,row in data.iterrows():
    row_text = '|'.join(f"Index:{idx}:{row[col]}" for col in data.columns if pd.notna(row[col]))

    context = context+ row_text + '\n'
```

```python
prompt = 'What is the maximum tips paid by male and female'

augmented_prompt = f'''
        <Role> You are a helpful assistant using RAG.
        <Goal> Answer the question asked by the user. Here are the questions {prompt}.
        <Context> Here are the documents retrieve from the Vector database to support the answer which you have to generate {context}
        '''
```

```python
response = model.generate_content(augmented_prompt)
print(response.text)
```
